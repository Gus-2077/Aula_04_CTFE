# Desafio: Média da Fiap


```python
# 1 Semestre

# CP - 1 Semestre
CP1, CP2, CP3 = input('Insira as notas das CP: ').split()
CP1 = float(CP1)
CP2 = float(CP2)
CP3 = float(CP3)

if CP1 > CP2 < CP3:
  CP_media = (CP1+CP2)/2
elif CP1 < CP2 > CP3:
  CP_media = (CP1+CP3)/2
else:
  CP_media = (CP2+CP3)/2

# SP - 1 Semestre
SP1, SP2 = input('Insira as notas das SP: ').split()
SP1 = float(SP1)
SP2 = float(SP2)

SP = (SP1 + SP2)/2

# Global Solution - 1 Semestre
gb = float(input('Insira a nota do GB: '))

media_semestral = (CP_media*0.2) + (gb*0.6) + (SP*0.2)
```

```python
# 2 Semestre

# CP - 2 Semestre
CP1_2, CP2_2, CP3_2 = input('Insira as notas das CP: ').split()
CP1_2 = float(CP1_2)
CP2_2 = float(CP2_2)
CP3_2 = float(CP3_2)

if CP1_2 > CP2_2 < CP3_2:
  CP_media_2 = (CP1_2+CP2_2)/2
elif CP1_2 < CP2_2 > CP3_2:
  CP_media_2 = (CP1_2+CP3_2)/2
else:
  CP_media_2 = (CP2_2+CP3_2)/2

# SP - 2 Semestre
SP1_2, SP2_2 = input('Insira as notas das SP: ').split()
SP1_2 = float(SP1_2)
SP2_2 = float(SP2_2)

SP_2 = (SP1_2 + SP2_2)/2

# Global Solution - 2 Semestre
gb_2 = float(input('Insira a nota do GB: '))

media_semestral_2 = (CP_media_2*0.2) + (gb_2*0.6) + (SP_2*0.2)
```

```python
# Media Final e Situação
media_final = (media_semestral*0.4) + (media_semestral_2*0.6)

if media_final >= 6:
  print('Aprovado')
  print(f'e sua média final foi: {media_final:.2f}')
# Exame
  print('Em Exame')
  nota_exame = float(input('Insira a nota do exame: '))
  media_final2 = (media_final + nota_exame)/2
  if media_final2 >= 6:
    print('Aprovado no exame')
  else:
    print('Reprovado no exame')
else:
  print('Reprovado')
  print(f'e sua média final foi: {media_final:.2f}')
```
