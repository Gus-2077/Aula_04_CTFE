# **Exemplo - Estatística Descritiva**


```python
A, B, C = input('').split()
A = float(A)
B = float(B)
C = float(C)
```

```python
PA = 2
PB = 3
PC = 5
```

```python
media = (A*PA + B*PB + C*PC)/(PA+PB+PC)

print(f'MEDIA = {media:.1f}')
```

# Exemplo - Cálculo de variância e desvio padrão

```python
n1, n2, n3, n4 = input('Numeros das distribuiçao: \n').split()
n1 = int(n1)
n2 = int(n2)
n3 = int(n3)
n4 = int(n4)

# Média
media = (n1+n2+n3+n4)/4

# Desvio
d1 = n1 - media
d2 = n2 - media
d3 = n3 - media
d4 = n4 - media

# Somatória do quadrado dos desvios
soma = d1**2 + d2**2 + d3**2 + d4**2
# Variância da distribuição
variancia = soma/4
# Desvio Padrão
desvio_padrao = variancia**0.5
print(f'MEDIA = {media:.2f}')
print(f'VARIANCIA = {variancia:.3f}')
print(f'DESVIO PADRAO = {desvio_padrao:.3f}')
```
