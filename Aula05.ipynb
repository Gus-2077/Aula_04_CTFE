```python
n1, n2, n3, n4 = map(float, input('Insira suas notas: ').split())

media = (n1*2+n2*3+n3*4+n4)/10

print(f'')
print(f'Media: {media:.1f} \n')

if media >= 7  :
  print('Aprovado')
elif media <= 6.9 and media >= 5:
  print('Em Exame \n')
  nota_exame = float(input('Insira sua nota do exame: '))
  print(f'Nota do exame: {nota_exame:.1f}')
  resultado = (media+nota_exame)/2
  if resultado >= 5:
    print('Aprovado')
    print(f'Media Final: {resultado:.1f}')
  else:
    print('Reprovado')
    print(f'Media Final: {resultado:.1f}')
else:
  print('Reprovado')
```
